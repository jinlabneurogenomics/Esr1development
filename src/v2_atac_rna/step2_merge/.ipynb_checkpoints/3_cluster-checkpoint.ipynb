## STEP 2: Clustering

```r
E_V.merged.atac.obj <- FindNeighbors(
  object = E_V.merged.atac.obj,
  reduction = 'lsi',
  dims = 2:30
)

E_V.merged.atac.obj
```

```r
Graphs(E_V.merged.atac.obj)
```

```r
E_V.merged.atac.obj <- FindClusters(
  object = E_V.merged.atac.obj,
  algorithm = 3,
  resolution = 0.8,
  verbose = FALSE,
  graph.name = 'ATAC_snn'
)

E_V.merged.atac.obj
```

## STEP 4: Compute cell type composition per cluster

```r
levels(E_V.merged.atac.obj$seurat_clusters)
```

```r
# Get most represented subclass per Leiden cluster

predominant_cluster_l <- list()
for (cluster_i in levels(E_V.merged.atac.obj$seurat_clusters)) {
    
    metadata <- E_V.merged.atac.obj@meta.data
    temp_metadata <- metadata[metadata$seurat_cluster == cluster_i, ]
    percentages_subclass_table <- table(temp_metadata$predicted.id) / sum(table(temp_metadata$predicted.id)) * 100
    predominant.id <- names(which.max(percentages_subclass_table))
    percentage_occupied <- percentages_subclass_table[[predominant.id]]

    predominant_cluster_l[[cluster_i]] <- c(cluster_i, predominant.id, percentage_occupied)
    
    }
    
# https://www.w3schools.com/r/r_for_loop.asp
# https://statisticsglobe.com/r-find-index-of-max-and-min-value-of-vector-or-data-frame-column
# https://stackoverflow.com/questions/21422188/how-to-get-name-from-a-value-in-an-r-vector-with-names
# https://stackoverflow.com/questions/15736719/how-do-i-extract-just-the-number-from-a-named-number-without-the-name
# https://www.geeksforgeeks.org/how-to-filter-r-dataframe-by-values-in-a-column/

# https://stackoverflow.com/questions/12511648/building-a-list-in-a-loop-in-r-getting-item-names-correct
# https://stackoverflow.com/questions/13442461/populating-a-data-frame-in-r-in-a-loop
```

```r
predominant_clusters_df <- data.frame(t(data.frame(predominant_cluster_l)))
colnames(predominant_clusters_df) <- c("cluster", "predicted.subclass", "percentage")
predominant_clusters_df
# https://sparkbyexamples.com/r-programming/convert-list-to-r-dataframe/
```
