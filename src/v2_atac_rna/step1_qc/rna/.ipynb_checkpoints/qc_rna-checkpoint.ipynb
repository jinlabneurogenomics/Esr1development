## Step 1: Load RNA data
- https://satijalab.org/seurat/reference/readmtx

```r
all_rna_mtx <- ReadMtx(
  mtx = all_rna_mtx_path, features = all_rna_features_path,
  cells = all_rna_barcodes_path
)
```

```r
gfp_rna_mtx <- ReadMtx(
  mtx = gfp_rna_mtx_path, features = gfp_rna_features_path,
  cells = gfp_rna_barcodes_path
)
```

```r
dim(all_rna_mtx)
head(colnames(all_rna_mtx))
head(rownames(all_rna_mtx))
```

```r
dim(gfp_rna_mtx)
head(colnames(gfp_rna_mtx))
head(rownames(gfp_rna_mtx))
```

## Step 2: Get combined counts matrix for all cells in both conditions (all, GFP+)

```r
all_rna_counts_subset_mtx <- all_rna_mtx[, all_rna_barcodes_no_prefix_l]
all_GFP_counts_subset_mtx <- gfp_rna_mtx[, gfp_barcodes_no_prefix_l]
```

```r
dim(all_rna_counts_subset_mtx)
```

```r
dim(all_GFP_counts_subset_mtx)
```

```r
head(colnames(all_rna_counts_subset_mtx))
```

```r
head(colnames(all_GFP_counts_subset_mtx))
```

### Prefix the column names of counts matrices
- https://stackoverflow.com/questions/24055308/add-a-prefix-to-all-rows-in-r

```r
colnames(all_rna_counts_subset_mtx) <- paste0('all_all_', colnames(all_rna_counts_subset_mtx))
colnames(all_GFP_counts_subset_mtx) <- paste0('gfp_GFP_', colnames(all_GFP_counts_subset_mtx))
```

```r
head(colnames(all_rna_counts_subset_mtx))
dim(all_rna_counts_subset_mtx)
```

```r
head(colnames(all_GFP_counts_subset_mtx))
dim(all_GFP_counts_subset_mtx)
```

```r
# concatenate the matrices horizontally

rna.combined.counts.mtx <- cbind(all_rna_counts_subset_mtx, all_GFP_counts_subset_mtx)

# https://stackoverflow.com/questions/7324833/concatenating-matrices-in-r
# https://www.rdocumentation.org/packages/base/versions/3.6.2/topics/cbind
```

```r
head(colnames(rna.combined.counts.mtx))
tail(colnames(rna.combined.counts.mtx))
dim(rna.combined.counts.mtx)
```

## Create ONE Seurat object

```r
annotated_rna_obj <- CreateSeuratObject(
    counts = rna.combined.counts.mtx,
    assay = "RNA",
    meta.data = annotated.meta.data
    
)
     
```

```r
annotated_rna_obj
```

```r
annotated_rna_obj@meta.data
```

## Step 3: processing (no relabeling, just to obtain UMAP)

### 3.1 SCTransform, PCA

```r
DefaultAssay(annotated_rna_obj) <- "RNA"
annotated_rna_obj <- SCTransform(annotated_rna_obj)
annotated_rna_obj <- RunPCA(annotated_rna_obj)
```

```r
options(repr.plot.width=10, repr.plot.height=5)
ElbowPlot(annotated_rna_obj)
annotated_rna_obj
```

### 3.2 Get RNA Neighbor Graph

```r
# build a neighbor graph using SCT assay
DefaultAssay(annotated_rna_obj) <- 'SCT'
annotated_rna_obj <- FindNeighbors(object = annotated_rna_obj,
                                  reduction = 'pca',
                                  dims = 1:50,
                                  verbose = TRUE
                                  )
```

```r
Graphs(annotated_rna_obj)
```

### 3.3 Get RNA UMAP

```r
annotated_rna_obj <- RunUMAP(
  object = annotated_rna_obj,
  reduction = "pca", 
  dims = 1:50,
  reduction.name = "rna_only_UMAP",
  reduction.key = "rna_only_UMAP_",
  verbose = TRUE
)
     
# https://satijalab.org/seurat/articles/integration_introduction.html
# https://satijalab.org/seurat/reference/runumap
```
