## STEP 2: Re-annotate object

```r
DefaultAssay(E_V.merged.atac.obj) <- 'ATAC'
```

```r
E_V.merged.atac.obj <- FindNeighbors(E_V.merged.atac.obj, reduction = 'lsi', dims = 2:30)
E_V.merged.atac.obj <- FindClusters(E_V.merged.atac.obj, algorithm = 3, resolution = 0.1)
new.cluster.ids <- c("L6 IT", "L6 CT", "Unknown 1", "Unknown 2")
names(new.cluster.ids) <- levels(E_V.merged.atac.obj)
E_V.merged.atac.obj <- RenameIdents(E_V.merged.atac.obj, new.cluster.ids)
E_V.merged.atac.obj$atac_clusters <- Idents(E_V.merged.atac.obj)
DimPlot(E_V.merged.atac.obj, group.by = 'atac_clusters','label'=TRUE)
```
