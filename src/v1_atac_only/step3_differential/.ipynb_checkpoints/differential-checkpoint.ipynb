# Differential on ALL cells in merged object

```r
all.combined
```

```r
DefaultAssay(all.combined) <- 'ATAC'
Idents(all.combined) <- 'condition'

da.peaks.all.cells <- FindMarkers(
    
            object = all.combined,
            min.pct = 0.05,
            test.use = 'LR',
            latent.vars = 'counts_in_macs_peaks',
            ident.1 = c('E1', 'E2'),
            ident.2 = c('V1'),
            assay = "ATAC",
            logfc.threshold = 0.01,
            pseudocount.use = 1e-04
    
)
```

```r
head(da.peaks.all.cells, n=20)
dim(da.peaks.all.cells)
```

```r
options(repr.plot.width=5, repr.plot.height=5)

Idents(all.combined) <- 'condition'
levels(all.combined) <- unique(all.combined$condition)

CoveragePlot(
  object = all.combined,
  region = rownames(da.peaks.all.cells)[1],
  extend.upstream = 1000,
  extend.downstream = 1000,
  ncol = 1
)
```

# [DEPRECATED] Differential accessibility ("neurons")

```r
# neurons
neurons <- c('Car3',
            'L2/3 IT CTX-1',
            'L2/3 IT CTX-2',
            'L4/5 IT CTX',
            'L5 IT CTX',
            'L5 NP CTX',
            'L5 PT CTX',
            'L6 CT CTX',
            'L6 IT CTX',
            'L6b CTX'
            )
```

```r
Idents(all.combined) <- 'predicted.id'
all.combined.neurons.only <- subset(x = all.combined, idents = neurons)
all.combined.neurons.only
```

```r
Idents(all.combined.neurons.only) <- 'condition'
head(Idents(all.combined.neurons.only))
```

# [DEPRECATED] "Neurons"-only differential accessibility

```r
DefaultAssay(all.combined.neurons.only) <- 'ATAC'

da_peaks.neurons.only <- FindMarkers(
    
            object = all.combined.neurons.only,
            min.pct = 0.05,
            test.use = 'LR',
            latent.vars = 'counts_in_macs_peaks',
            ident.1 = c('E1', 'E2'),
            ident.2 = c('V1'),
            assay = "ATAC",
            logfc.threshold = 0.01,
            pseudocount.use = 1e-04
    
)

head(da_peaks.neurons.only)
```

```r
dim(da_peaks.neurons.only)
```

```r
da_peaks_fc_sorted <- da_peaks[order(da_peaks.neurons.only$avg_log2FC, decreasing = TRUE), ]

head(da_peaks_fc_sorted)
```
