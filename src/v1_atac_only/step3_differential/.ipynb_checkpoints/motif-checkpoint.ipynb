# Differential motif enrichment
https://stuartlab.org/signac/articles/motif_vignette.html

```r
all.combined <- RegionStats(
  object = all.combined,
  genome = BSgenome.Mmusculus.UCSC.mm10
)
```

```r
# Get a list of motif position frequency matrices from the JASPAR database
pfm <- getMatrixSet(
  x = JASPAR2022,
  opts = list(collection = "CORE", tax_group = 'vertebrates', all_versions = FALSE)
)
# https://stuartlab.org/signac/articles/motif_vignette.html
# https://stuartlab.org/signac/articles/data_structures.html#the-motif-class
```

```r
# add motif information
all.combined <- AddMotifs(
  object = all.combined,
  genome = BSgenome.Mmusculus.UCSC.mm10,
  pfm = pfm
)
```

## Progenitors
- Hes1: MA1099.1 (mouse), MA1099.2 (human)
- Nr2f1: MA0017.2 (human)
- Pax6: MA0069.1 (human)
- Lhx2: MA0700.2 (human)
- Emx1: MA0612.2 (human)
- Emx2: MA0886.1 (human) (basically identical to Emx1)
- Hes5: MA0821.2 (human)
- Tcf3: MA0522.1 (mouse), MA0522.3 (human)
- Sox9: MA0077.1 (human)
- Id4: NA (in JASPAR 2022)
- Sox2: MA0143.4 (human), MA0143.3 (mouse)
- Jun: MA0489.2 (mouse)
- Nr2e1: MA0676.1 (mouse)


```r
motifs.ids <- c('MA1099.2', 'MA0017.2', 'MA0069.1', 'MA0700.2',
              'MA0612.2', 'MA0886.1', 'MA0821.2', 'MA0522.3',
              'MA0077.1', 'MA0143.4', 'MA0489.2', 'MA0676.1')

motif.names <- c('Hes1', 'Nr2f1', 'Pax6', 'Lhx2', 'Emx1',
                'Emx2', 'Hes5', 'Tcf3', 'Sox9', 'Sox2',
                'Jun','Nr2e1')

gg_Fig <- FeaturePlot(
  object = all.combined,
  features = motifs.ids,
  min.cutoff = 'q10',
  max.cutoff = 'q90',
  pt.size = 0.1
)

for (i in 1: length(motifs.ids)) gg_Fig[[i]]$labels$title = motif.names[i]
gg_Fig
    
# https://github.com/satijalab/seurat/issues/1430
```

# Differential motifs?

```r
DefaultAssay(all.combined) <- 'chromvar'
Idents(all.combined) <- 'condition'

differential.activity <- FindMarkers(
  object = all.combined,
  ident.1 = c('E1', 'E2'),
  ident.2 = 'V1',
  only.pos = TRUE,
  mean.fxn = rowMeans,
  fc.name = "avg_diff"
)
```

```r
MotifPlot(
  object = all.combined,
  motifs = head(rownames(differential.activity)),
  assay = 'ATAC'
)
```
